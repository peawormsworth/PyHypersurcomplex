# Hypersurcomplex Nilpotent
A nilpotent is a number that square to 0. Zero itself is a simple nilpotent.

# Real Split-Quaternion Idempotent
The split-quaternion idempotent is proven below.

```python
from involution.algebra import SplitQuaternion
algebra = SplitQuaternion
print("Test Split Quaternion nilpotent - a number whose square is one")
for q in [
    algebra([0,1, 1, 0]),
    algebra([0,1,-1, 0]),
    algebra([0,1, 0, 1]),
    algebra([0,1, 0,-1]),
]:
    calc = q*q
    expect = algebra([0,0,0,0])
    print("\n       q = %r" % (q))
    print("\n     q×q = %r" % (calc))
    print("""

=== Real Quaternion Nilpotent test

Given the {0}:

          q = {1}

...verify it squares to itself:

        q×q = {2}
     expect = {3}

""".format(algebra.__name__, q, calc, expect))
    assert str(calc) == str(expect)
    
print('\nresult: success\\done.\n')
```
